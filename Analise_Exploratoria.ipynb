```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('dados.csv', sep='|')
df
```

```python
def converter_data(data):
    data = data[:10]
    data = data.replace('/','')
    data = datetime(int(data[4:]) , int(data[:2]) ,int(data[2:4]))
    return data
```

```python
df['Abertura'] = df['Abertura'].apply(converter_data)
df
```
